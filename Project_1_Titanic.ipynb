```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
# Asigne los datos a una una variable que llamare Titanic meadiante la funcion de Seaborn load_dataset
titanic=seaborn.load_dataset('titanic')
```

```python
# Se creo folder para guardar los datos en formato .csv mediante la funcion de pandas to_cvs
titanic.to_csv(r"data/df_titanic.csv")
```

```python
# Vemos el numero de filas y columnas de nuestro df llamado Titanic mediante la funcion de pandas shape
titanic.shape
```

```python
# Revisamos el porcentaje de columnas vacias
titanic.isna().sum()/titanic.shape[0]
```

```python
#Para cumplir con la variable categorica, vamos a calcular promedio (meadiante mean) de supervivencia por género
survival_by_sex=titanic.groupby("sex")["survived"].mean()

# Gráfica de barras con seleccion de color que se asemeje al sexo al que pertenece
survival_by_sex.plot(kind="bar", color=["pink", "blue"])

# Personalización de tamano y titulos en el nombre de la grafica, x y y
plt.title("Porcentaje de Supervivencia por Género", fontsize=14)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Tasa de Supervivencia", fontsize=12)

#establecemos limites (el limite de X se me hizo innecesario aqui pero para cumplir con el requerimiento jaja)
plt.ylim(0, 1)
plt.xlim(-1, 2)
# Giramos categorias de x
plt.xticks(rotation=0)
# Le damos mejor visibilidad a las cantidades
plt.grid(axis='y')

plt.show()
```

```python
# Histograma de tarifas (numérico) Calculé cuanto pagaron los pasajeros por su boleto, incluyo bins para segmentar, como parte de perzonalización
plt.hist(titanic["fare"], bins=10, color='mediumseagreen', edgecolor='black')
plt.title("Costos de boletos pagados por los pasajeros", fontsize=14)
plt.xlabel("Costo $", fontsize=12)
plt.ylabel("Cantidad de Pasajeros", fontsize=12)
plt.xlim(0, 500)   # Limita para evitar valores extremos que aplasten el gráfico
plt.ylim(0, 750)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y')

plt.show()
```
